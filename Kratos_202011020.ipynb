Overview: </p>
Recommendation Systems are a type of information filtering systems
as they improve the quality of search results and provide items that are more relevant to the search
item or are related to the search history of the user. They are used to predict the rating or
preference that a user would give to an item. </p> 
A movie recommendation system will recommend a user movie/s based on his previous responses for a particular genre, ratings, etc and on the basis of popularity a movie has amoung all the users i.e. collaborative recommendation. 

Explaining the project: 

• First of all, we get a dataset of movie ratings and make sure that we understand its structure.

• Next, we will try to get non-personalized recommendations followed by personalized.

• Then, at last, we will train our model to get a User-User collaborative filtering to get recommendations for combined users.

```python
# the generic recommendation which I got above doesn't seem to be much releavant
# here I am adding count of users who rated the movie with the particular rating

average_ratings = (data.ratings).groupby('item') \
       .agg(count=('user', 'size'), rating=('rating', 'mean')) \
       .reset_index()

sorted_avg_ratings = average_ratings.sort_values(by="rating", ascending=False)
joined_data = sorted_avg_ratings.join(data.movies['genres'], on='item')
joined_data = joined_data.join(data.movies['title'], on='item')
joined_data = joined_data[joined_data.columns[1:]]


print("RECOMMENDED FOR ALL USERS:")
joined_data.head(rows_to_show)
```

```python
# to improve my results for generic recommendation of movies I have set a minimum count of person
# those should have rated the movie, then only the movie will appear in the list below
minimum_to_include = 100

average_ratings = (data.ratings).groupby(['item']).mean()
rating_counts = (data.ratings).groupby(['item']).count()
average_ratings = average_ratings.loc[rating_counts['rating'] > minimum_to_include]
sorted_avg_ratings = average_ratings.sort_values(by="rating", ascending=False)
joined_data = sorted_avg_ratings.join(data.movies['genres'], on='item')
joined_data = joined_data.join(data.movies['title'], on='item')
joined_data = joined_data[joined_data.columns[3:]]

print("RECOMMENDED FOR ALL USERS:")
joined_data.head(rows_to_show)
# results improves :))
```

```python
# we have recommended the movies based on the ratings 
# now we will narrow down our recommendations based on particular genre

average_ratings = (data.ratings).groupby(['item']).mean()
rating_counts = (data.ratings).groupby(['item']).count()
average_ratings = average_ratings.loc[rating_counts['rating'] > minimum_to_include]
average_ratings = average_ratings.join(data.movies['genres'], on='item')
# we have filtered the recommendations on the basis of a perticular genre, here 'Action'
average_ratings = average_ratings.loc[average_ratings['genres'].str.contains('Action')]

sorted_avg_ratings = average_ratings.sort_values(by="rating", ascending=False)
joined_data = sorted_avg_ratings.join(data.movies['title'], on='item')
joined_data = joined_data[joined_data.columns[3:]]
print("RECOMMENDED MOVIES FOR AN ACTION MOVIE FAN:")
joined_data.head(rows_to_show)
```

We have seen generic recommendations based on all users ratings narrowed down to specific genres, now we will personalized the recommendations by using personal watch data, personal ratings, etc. </p> 

Moreover, if we notice till now we will get the same recommendations even if we have watched the moive. So further, I will use personal watch data set to get new recommendations.

```python
import csv

personal_rating_1 = {}
personal_rating_2 = {}

with open("/content/Movie-Recommendation-System/personal_rating_1.csv", newline='') as csvfile:
  ratings_reader = csv.DictReader(csvfile)
  for row in ratings_reader:
    if ((row['ratings'] != "") and (float(row['ratings']) > 0) and (float(row['ratings']) < 6)):
      personal_rating_1.update({int(row['item']): float(row['ratings'])})
      
with open("/content/Movie-Recommendation-System/personal_rating_2.csv", newline='') as csvfile:
  ratings_reader = csv.DictReader(csvfile)
  for row in ratings_reader:
    if ((row['ratings'] != "") and (float(row['ratings']) > 0) and (float(row['ratings']) < 6)):
      personal_rating_2.update({int(row['item']): float(row['ratings'])})
     
print("Rating dictionaries assembled!")
print("Sanity check:")
print("\tPersonal rating 1 for 1 (Toy Story) is " + str(personal_rating_1[1]))
print("\tPersonal rating 2 for 1 (Toy Story) is " + str(personal_rating_2[1]))
```

```python
# now we will train a new collaborative filtering model to provide recommendations

from lenskit.algorithms import Recommender
from lenskit.algorithms.user_knn import UserUser
#number of recommendations is set to 10, can be modified
num_recs = 10  

#These two numbers set the minimum (3) and maximum (15) number of neighbors to consider. 
user_user = UserUser(15, min_nbrs=3) 
algo = Recommender.adapt(user_user)
algo.fit(data.ratings)

print("Set up a User-User algorithm!")
```

```python
#Here, -1 tells it that it's not an existing user in the set, that we're giving new ratings, while 10 is how many recommendations it should generate
pers_recs_1 = algo.recommend(-1, num_recs, ratings=pd.Series(personal_rating_1)) 

joined_data = pers_recs_1.join(data.movies['genres'], on='item')      
joined_data = joined_data.join(data.movies['title'], on='item')
joined_data = joined_data[joined_data.columns[2:]]
print("\n\nFirst Personalized Recommendations:")
joined_data
```

```python
#Here, -1 tells it that it's not an existing user in the set, that we're giving new ratings, while 10 is how many recommendations it should generate
pers_recs_2 = algo.recommend(-1, num_recs, ratings=pd.Series(personal_rating_2)) 

joined_data = pers_recs_2.join(data.movies['genres'], on='item')      
joined_data = joined_data.join(data.movies['title'], on='item')
joined_data = joined_data[joined_data.columns[2:]]
print("Second Personal Recommendations:")
joined_data
```

• We have successfully provided personalized recommendations for movies to different users. </p> 

• Now we can take our recommedations to one step further. We will combine recommendations of 2 or more users to provide them a common recommendation. </p>

• This feauture can be helpful for a movie night when more than 1 person wants to watch a movie. Using this feature they will be able to get a recommendaion based of watch history of each one of them.

</p> Let's Do IT :))

```python
combined_rating_dict = {}
for k in personal_rating_1:
  if k in personal_rating_2:
    combined_rating_dict.update({k: float((personal_rating_1[k]+personal_rating_2[k])/2)})
  else:
    combined_rating_dict.update({k:personal_rating_1[k]})
for k in personal_rating_2:
   if k not in combined_rating_dict:
      combined_rating_dict.update({k:personal_rating_2[k]})
      
print("Combined ratings dictionary assembled!")
print("Sanity check:")
print("\tCombined rating for 1 (Toy Story) is " + str(combined_rating_dict[1]))
```

```python
# now finally we can use our combine dictionary to get a common recommendation for group
combined_recs = algo.recommend(-1, num_recs, ratings=pd.Series(combined_rating_dict))  
#Here, -1 tells it that it's not an existing user in the set, that we're giving new ratings, while 10 is how many recommendations it should generate
joined_data = combined_recs.join(data.movies['genres'], on='item')      
joined_data = joined_data.join(data.movies['title'], on='item')
joined_data = joined_data[joined_data.columns[2:]]
print("\n\nPersonalized Recommendation for user 1 and user 2:")
joined_data
```
